```python
def rbf_1d(xx, cc, hh):
    return np.exp(-(xx-cc)**2/hh**2)

def sigmoid(xx, vv, bb):
    return 1/(1+np.exp(-vv*xx-bb))
```

```python
plt.clf()
grid_size = 0.01
x_grid = np.arange(-10, 10, grid_size)
plt.plot(x_grid, rbf_1d(x_grid, cc=5, hh=1), '-b')
plt.plot(x_grid, rbf_1d(x_grid, cc=-2, hh=2), '-r')
plt.show()
```

```python
def phi_linear(Xin):
    return np.dstack([np.ones(len(Xin)), Xin])[0]

def phi_quadratic(Xorig):
    return np.dstack([np.ones(len(Xorig)), Xorig, Xorig**2])[0]

def phi_cubic(Xorig):
    return np.dstack([np.ones(len(Xorig)), Xorig, Xorig**2, Xorig**3])[0]

def fw_rbf(xx, cc):
    return np.exp(-(xx-cc)**2)/2

def phi_rbf(Xin):
    return np.dstack([fw_rbf(Xin, 1), fw_rbf(Xin, 2), fw_rbf(Xin, 3)])[0]

def fit_and_plot(phi_fn, X, yy):
    plt.clf()
    w_fit = np.linalg.lstsq(phi_fn(X), yy)[0]
#     print(w_fit.shape)
    X_grid = np.arange(0, 4, 0.01)
#     print(X_grid.shape)
    f_grid = np.dot(phi_fn(X_grid), w_fit)
#     print(f_grid.shape)
    plt.plot(X_grid, f_grid, '-r')
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = np.asarray([i for i in range(1,11)])
y = np.asarray([7, 8, 7, 13, 16, 15, 19, 23, 18, 21])
X = X.reshape(10, 1)
y = y.reshape(10, 1)
```

```python
plt.plot(X, y, 'ro')
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
a = model.coef_ * X + model.intercept_
```

```python
plt.plot(X, y, 'ro', X, a)
axes = plt.gca()
axes.set_ylim([0, 30])
plt.show()
```

```python
print(model.score(X, y))
print(X)
```
